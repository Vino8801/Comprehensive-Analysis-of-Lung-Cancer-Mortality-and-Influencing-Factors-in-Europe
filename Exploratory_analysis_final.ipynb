```python
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/lung cancer.csv")
print(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the occurrences of each cancer stage
cancer_stage_counts = data['cancer_stage'].value_counts()

# Define labels for the different cancer stage
labels = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4']

# Plotting the pie chart
plt.figure(figsize=(4, 4))
plt.pie(cancer_stage_counts, labels=labels, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Distribution of Cancer stages')

# Displaying the plot
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
cross = pd.crosstab(data["cancer_stage"], data["survived"])
sns.heatmap(cross, annot=True, fmt='d')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='cancer_stage', hue='survived')
plt.title('Survival based on stage Status')
plt.xlabel('Cancer stage status')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
ct = data['country'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(data['country'].value_counts(), labels=ct.index, autopct='%1.1f%%', startangle=140)
plt.title('Country Distribution')
plt.show()
```

```python
# Grouping data by country and survival status
grouped_data = data.groupby(['country', 'survived']).size().unstack()

# Plotting the bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Survival Count by Country')
plt.legend(title='Survived', labels=['Not Survived (0)', 'Survived (1)'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Printing the grouped data (value count by country and survival status)
print(grouped_data)
```

```python
# Grouping data by country and survival status
grouped_data = data.groupby(['country', 'survived']).size().unstack(fill_value=0)

# Sorting the data to get the top 5 countries with the highest count of 0 (not survived) and 1 (survived)
top_10_not_survived = grouped_data[0].sort_values(ascending=False).head(10)
top_10_survived = grouped_data[1].sort_values(ascending=False).head(10)

# Plotting the bar chart for top 5 countries with highest count of 0 (not survived)
plt.figure(figsize=(8, 4))
top_10_not_survived.plot(kind='bar', color='red', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries with Highest Count of Not Survived (0)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting the bar chart for top 5 countries with highest count of 1 (survived)
plt.figure(figsize=(8, 4))
top_10_survived.plot(kind='bar', color='green', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries with Highest Count of Survived (1)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
